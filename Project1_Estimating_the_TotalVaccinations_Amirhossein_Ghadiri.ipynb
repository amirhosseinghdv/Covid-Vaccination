## Now, since we can observe that total_vaccinations approximately equals people_vaccinated plus people_fully_vaccinated, we handle other rows based on this concept.

```python
df = df.reset_index(drop=True)
df
```

```python
for i in range(df.shape[0]):
    if (not pd.isnull(df.iloc[i, 2])) and (not pd.isnull(df.iloc[i, 3])) and pd.isnull(df.iloc[i, 4]):
        df.iloc[i,4] = df.iloc[i, 2] - df.iloc[i, 3]
```

```python
df.info()
```

```python
for i in range(df.shape[0]):
    if pd.isnull(df.iloc[i, 2]) and (not pd.isnull(df.iloc[i, 3])) and (not pd.isnull(df.iloc[i, 4])):
        df.iloc[i, 2] = df.iloc[i, 3] + df.iloc[i,4]
```

```python
df.info()
```

```python
for i in range(df.shape[0]):
    if (not pd.isnull(df.iloc[i, 2])) and pd.isnull(df.iloc[i, 3]) and (not pd.isnull(df.iloc[i, 4])):
        df.iloc[i, 3] = df.iloc[i, 2] - df.iloc[i,4]
```

## In order to calculate total_vaccinations_per_hundred, people_vaccinated_per_hundred, and people_fully_vaccinated_per_hundred, we can exploit the fact that the population for each country remains the same; consequently, there exists a particular j index for each country.

```python
d1={}
for i in range(df.shape[0]):
    country = df.iloc[i,0]
    if (not pd.isnull(df.iloc[i, 2])) and (not pd.isnull(df.iloc[i, 7])):
        index = df.iloc[i, 2] / df.iloc[i, 7]
        d1[country]=index
        
```

```python
len(d1.keys())
```

```python
df.country.unique().shape
```

## Now that we have found out the index for each country, we can calculate total_vaccinations_per_hundred, people_vaccinated_per_hundred, and people_fully_vaccinated_per_hundred by deviding total_vaccinations, people_vaccinated, and people_fully_vaccinated by the index for each country.

```python
for i in range(df.shape[0]):
    country = df.iloc[i,0]
    if  pd.isnull(df.iloc[i, 7]):
        df.iloc[i, 7] = df.iloc[i, 2] / d1[country]
    if  pd.isnull(df.iloc[i, 8]):
        df.iloc[i, 8] = df.iloc[i, 3] / d1[country]
    if  pd.isnull(df.iloc[i, 9]):
        df.iloc[i, 9] = df.iloc[i, 4] / d1[country]        
        
```

# Train your model

```python
# for 10 trees
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(x_train,y_train)
```

```python
rfr_pred_train = regressor.predict(x_train)
```

```python
rfr_pred_train.shape
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
def MAPE(Y_actual,Y_Predicted):
    mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_actual))*100
    return mape
```

```python
MAPE(y_train, rfr_pred_train)
```

```python
rfr_pred_test = regressor.predict(x_test)
```

```python
MAPE(y_test, rfr_pred_test)
```
